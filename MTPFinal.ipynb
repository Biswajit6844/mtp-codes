```python
#Incorporatig Optimized  Swipt
import numpy as np
import pandas as pd

N = 800
num_time_slots = 2
T = 60
num_intervals_per_slot = 10
total_transmit_power = 1.0

max_m, max_n = 20, 40
# Arrays to store the energy harvested and transmitted for each interval
interval_data = {
    'Time Slot': [],
    'Interval': [],
    'Energy Harvested (J)': [],
    'Energy Transmitted (J)': [],
    'm': [],
    'n': []
}

def tau(t):

    return 0.5 + 0.5 * (-1)**t

for time_slot in range(1, num_time_slots + 1):
    for interval in range(1, num_intervals_per_slot + 1):

        m = np.random.randint(1, max_m + 1)
        n = np.random.randint(1, max_n + 1)

        while m * n > N:
            m = np.random.randint(1, max_m + 1)
            n = np.random.randint(1, max_n + 1)


        lambda_harvest = m * n / N
        tau_current = tau(interval)


        harvested_energy = lambda_harvest * total_transmit_power * tau_current * (T / num_intervals_per_slot)
        transmitted_energy = (1 - lambda_harvest) * total_transmit_power * (1 - tau_current) * (T / num_intervals_per_slot)


        interval_data['Time Slot'].append(time_slot)
        interval_data['Interval'].append(interval)
        interval_data['Energy Harvested (J)'].append(harvested_energy)
        interval_data['Energy Transmitted (J)'].append(transmitted_energy)
        interval_data['m'].append(m)
        interval_data['n'].append(n)

# Convert interval data to a DataFrame and display
df_intervals = pd.DataFrame(interval_data)
print(df_intervals.to_string(index=False))

# Calculate and display the total energy for each time slot
total_energy = df_intervals.groupby('Time Slot').sum()
print("\nTotal Energy for Each Time Slot:")
print(total_energy[['Energy Harvested (J)', 'Energy Transmitted (J)']].to_string(index=True))
```

```python
N = 800
num_time_slots = 2
T = 60
num_intervals_per_slot = 10
total_transmit_power = 1.0


max_m, max_n = 20, 40


interval_data = []

def tau(t):

    return 0.5 + 0.5 * (-1)**t


phase_shifts = np.exp(1j * 2 * np.pi * np.random.rand(N))


for time_slot in range(1, num_time_slots + 1):
    for interval in range(1, num_intervals_per_slot + 1):

        m = np.random.randint(1, max_m + 1)
        n = np.random.randint(1, max_n + 1)
        # Ensure m * n does not exceed N
        while m * n > N:
            m = np.random.randint(1, max_m + 1)
            n = np.random.randint(1, max_n + 1)


        lambda_harvest = m * n / N
        tau_current = tau(interval)




        efficiency_factor = np.abs(np.sum(phase_shifts[:m * n])) / (m * n)  # Normalized efficiency due to phase alignment


        harvested_energy = efficiency_factor * lambda_harvest * total_transmit_power * tau_current * (T / num_intervals_per_slot)
        transmitted_energy = (1 - lambda_harvest) * total_transmit_power * (1 - tau_current) * (T / num_intervals_per_slot)

        # Store interval data
        interval_data.append({
            'Time Slot': time_slot,
            'Interval': interval,
            'm': m,
            'n': n,
            'Efficiency Factor': efficiency_factor,
            'Energy Harvested (J)': harvested_energy,
            'Energy Transmitted (J)': transmitted_energy,
        })

# Convert interval data to a DataFrame and display
df_intervals = pd.DataFrame(interval_data)
df_intervals[['Time Slot', 'Interval', 'm', 'n', 'Efficiency Factor', 'Energy Harvested (J)', 'Energy Transmitted (J)']]
```

```python

import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Dense, Input, BatchNormalization, Activation
from tensorflow.keras.optimizers import Adam
import random
import collections
from tabulate import tabulate

class CommunicationEnvironment:
    def __init__(self, max_steps):

        self.sinr_history = []
        self.efficiency_history = []
        self.stability_check_window = 10


        self.current_step = 0
        self.max_steps = max_steps

        self.H1 = np.random.rand(1, 1)
        self.h2k = np.random.rand(800, 3)


        self.R = np.random.uniform(0, 30)
        self.alpha = np.random.uniform(2, 5)
        self.beta = np.random.uniform(0, 1)


        self.M = 1
        self.N = 800
        self.K = 3
        self.power_for_energy_harvesting = np.zeros(self.K)



        self.Phi = np.exp(1j * np.random.rand(self.N) * 2 * np.pi)
        self.wi = np.random.rand(self.M, self.K)
        self.x = np.random.rand(self.M, 1)
        self.sk = np.random.rand(self.K, 1)
        self.sigma = 10**(-8/10)
        self.gamma_k = np.random.rand(self.K)
        self.B = 1e6
        self.zeta = 1.2
        self.P_BS = 10**(39/10)
        self.P_k = np.random.rand(self.K) * 10**(30/10)
        self.P_UAV = 10**(30/10)
        self.P_UE = 10**(10/10)
        self.P_RIS = self.N * 10**(0.8/10)
        self.P_RISn = 10**(0.8/10)


        self.L_BS = np.array([-200, 0, 0])
        self.Q_j = np.array([100, 50, 100])
        self.user_positions = np.array([[75, 25, 0], [125, 25, 0], [100, 75, 0]])
        self.obstacle_position = np.array([80, 40, 0])


    def reset(self):

        self.H1 = np.random.rand(1, 1)
        self.h2k = np.random.rand(800, 3)


        self.Q_j = np.array([100, 50, 100])



        self.Phi = np.exp(1j * np.random.rand(self.N) * 2 * np.pi)
        self.wi = np.random.rand(self.M, self.K)
        self.x = np.random.rand(self.M, 1)
        self.sk = np.random.rand(self.K, 1)
        self.gamma_k = np.random.rand(self.K)

        # Return the initial state of the environment

        state = self._get_state()
        return state

    def _get_state(self):
        # This method should return the current state of the environment


        state = np.concatenate((self.H1.flatten(), self.h2k.flatten(), self.Q_j, self.user_positions.flatten()))
        return state


    def step(self, action):

        # Apply the action to the environment
        self._apply_action(action)
        # Update the environment state based on the action
        previous_energy_harvested = self.get_energy_harvested()
        self._update_environment()

        # Adjust UAV position to increase energy harvested
        self.adjust_uav_position_for_optimal_energy_harvesting()



        # Calculate the reward based on the new state
        reward = self._calculate_reward()

        # Check if the episode is done
        done = self._check_done()

        # Get the new state
        new_state = self._get_state()

        return new_state, reward, done


    def adjust_uav_position_for_optimal_energy_harvesting(self):
        # Optimization algorithm to find the best UAV position for maximizing the objective
        best_position = self.Q_j[:2]
        best_energy = self.get_energy_harvested()
        best_h2k = np.copy(self.h2k)

        for _ in range(10):
            trial_position = self.Q_j[:2] + (np.random.rand(2) - 0.5) * 10
            trial_position = self._apply_position_constraints(trial_position)


            original_position = np.copy(self.Q_j)
            self.Q_j[:2] = trial_position


            self._update_environment()


            trial_energy = self.get_energy_harvested()


            if trial_energy > best_energy:
                best_position = trial_position
                best_energy = trial_energy
                best_h2k = np.copy(self.h2k)


            self.Q_j = original_position


        self.Q_j[:2] = best_position
        self.h2k = best_h2k
        self._update_environment()










    def _apply_action(self, action):
      uav_movement = action[:2]
      phase_shift_changes = action[2:]
      new_position = self.Q_j[:2] + uav_movement
      self.Q_j[:2] = self._apply_position_constraints(new_position)


      if len(phase_shift_changes) < len(self.Phi):
        phase_shift_changes = np.concatenate([phase_shift_changes, np.zeros(len(self.Phi) - len(phase_shift_changes))])
      elif len(phase_shift_changes) > len(self.Phi):
        phase_shift_changes = phase_shift_changes[:len(self.Phi)]


      phase_shift_changes = np.array(phase_shift_changes)


      current_angles = np.angle(self.Phi)


      new_angles = (current_angles + phase_shift_changes) % (2 * np.pi)


      self.Phi = np.abs(self.Phi) * np.exp(1j * new_angles)









    def _apply_position_constraints(self, position):


        max_x, max_y = 200, 200
        min_x, min_y = -200, -200
        constrained_x = np.clip(position[0], min_x, max_x)
        constrained_y = np.clip(position[1], min_y, max_y)
        return np.array([constrained_x, constrained_y])

    def _update_environment(self):

        distance_bs_ris = np.linalg.norm(self.L_BS - self.Q_j)
        self.H1 = self._calculate_path_loss(distance_bs_ris)


        for k in range(self.K):
            distance_ris_user = np.linalg.norm(self.user_positions[k][:2] - self.Q_j[:2])
            self.h2k[:, k] = self._calculate_path_loss(distance_ris_user) * self._apply_phase_shifts(self.Phi)


        self._update_swipt_signals()










    def _calculate_path_loss(self, distance):
      frequency = 2.4e9
      c = 3e8
      path_loss_0 = 20 * np.log10(4 * np.pi * frequency / c)

      path_loss = 20 * np.log10(distance) + path_loss_0
      path_loss_linear = 10 ** (-path_loss / 10)

      return path_loss_linear



    def _apply_phase_shifts(self, phi):

        return np.exp(1j * phi)

    def _update_swipt_signals(self):



        power_split_ratio = 0.5
        self.power_for_information = self.P_k * power_split_ratio
        self.power_for_energy_harvesting = self.P_k * (1 - power_split_ratio)


    def _calculate_reward(self):

        reward_signal_quality = self._calculate_signal_quality()
        reward_energy_efficiency = self._calculate_energy_efficiency()
        total_reward = reward_signal_quality + reward_energy_efficiency
        return total_reward

    def _calculate_signal_quality(self):


        sinr_values = []

        for k in range(self.K):


            signal_power = np.abs(self.h2k[:, k])**2


            interference_power = 1


            noise_power = self.sigma


            sinr = signal_power / (interference_power + noise_power)
            sinr_values.append(sinr)


        average_sinr = np.mean(sinr_values)


        signal_quality_reward = np.log(1 + average_sinr)  # Log-scale reward

        self.sinr_history.append(signal_quality_reward)


        return signal_quality_reward


    def _calculate_energy_efficiency(self):

        total_energy_harvested = np.sum(self.power_for_energy_harvesting)


        total_energy_used = np.sum(self.power_for_information)


        if total_energy_used > 0:
            energy_efficiency = total_energy_harvested / total_energy_used
        else:
            energy_efficiency = 0


        energy_efficiency_reward = energy_efficiency

        self.efficiency_history.append(energy_efficiency_reward)

        return energy_efficiency_reward

    def _check_done(self):

        self.current_step += 1
        if self.current_step >= self.max_steps:
            return True
        return False

    def _check_stability_or_convergence(self):



        sinr_threshold = 0.01
        efficiency_threshold = 0.01


        if len(self.sinr_history) >= self.stability_check_window:
            recent_sinr_changes = np.abs(np.diff(self.sinr_history[-self.stability_check_window:]))
            recent_efficiency_changes = np.abs(np.diff(self.efficiency_history[-self.stability_check_window:]))


            if np.all(recent_sinr_changes < sinr_threshold) and np.all(recent_efficiency_changes < efficiency_threshold):
                return True

        return False

    def get_energy_harvested(self):

        return np.sum(self.power_for_energy_harvesting)

    def get_transmit_power(self):

        return self.P_k

    def get_uav_position(self):

        return self.Q_j

    def get_phase_shifts(self):

        return np.angle(self.Phi)




# Define DDPG components
class Actor(Model):

    def __init__(self, state_size, action_size, learning_rate=0.001):
        super(Actor, self).__init__()
        self.state_size = state_size
        self.action_size = action_size
        self.learning_rate = learning_rate


        self.dense1 = Dense(400, input_dim=self.state_size, activation='relu')
        self.dense2 = Dense(300, activation='relu')
        self.output_layer = Dense(self.action_size, activation='tanh')

        self.compile(optimizer=Adam(learning_rate=self.learning_rate))

    def call(self, state):
        x = self.dense1(state)
        x = self.dense2(x)
        action = self.output_layer(x)
        return action



class Critic(Model):
    # Define your critic network
    def __init__(self, state_size, action_size, learning_rate=0.001):
        super(Critic, self).__init__()
        self.state_size = state_size
        self.action_size = action_size
        self.learning_rate = learning_rate

        # State as input
        self.state_input = Input(shape=(self.state_size,))
        self.state_h1 = Dense(400, activation='relu')(self.state_input)
        self.state_h2 = Dense(300, activation='relu')(self.state_h1)

        # Action as input
        self.action_input = Input(shape=(self.action_size,))
        self.action_h1 = Dense(300, activation='relu')(self.action_input)

        # Combine state and action
        merged = tf.keras.layers.Concatenate()([self.state_h2, self.action_h1])

        # Final output layer
        self.output_layer = Dense(1, activation='linear')(merged)

        self.model = Model(inputs=[self.state_input, self.action_input], outputs=self.output_layer)
        self.optimizer = Adam(learning_rate=self.learning_rate)
        self.model.compile(loss='mse', optimizer=self.optimizer)

    def call(self, inputs):
        return self.model(inputs)

class ReplayBuffer:
    def __init__(self, buffer_size):
        self.buffer_size = buffer_size
        self.buffer = collections.deque(maxlen=buffer_size)

    def add(self, state, action, reward, next_state, done):
        experience = (state, action, reward, next_state, done)
        self.buffer.append(experience)

    def sample(self, batch_size):
        batch = random.sample(self.buffer, min(len(self.buffer), batch_size))
        states, actions, rewards, next_states, dones = map(np.stack, zip(*batch))
        return states, actions, rewards, next_states, dones

class DDPGAgent:
    def __init__(self, state_size, action_size, actor_lr, critic_lr, buffer_size, batch_size, tau=0.001):
        self.state_size = state_size
        self.action_size = action_size
        self.buffer_size = buffer_size
        self.batch_size = batch_size
        self.tau = tau


        self.actor = Actor(state_size, action_size, actor_lr)
        self.critic = Critic(state_size, action_size, critic_lr)


        self.target_actor = Actor(state_size, action_size, actor_lr)
        self.target_critic = Critic(state_size, action_size, critic_lr)


        self.replay_buffer = ReplayBuffer(buffer_size)


        self.update_target_networks()

    def update_target_networks(self):

        self._soft_update(self.actor, self.target_actor)
        self._soft_update(self.critic, self.target_critic)

    def _soft_update(self, model, target_model):
        for target_param, param in zip(target_model.trainable_variables, model.trainable_variables):
            target_param.assign(self.tau * param + (1 - self.tau) * target_param)

    def learn(self):

        states, actions, rewards, next_states, dones = self.replay_buffer.sample(self.batch_size)


        target_actions = self.target_actor(states)
        target_critic_values = self.target_critic([next_states, target_actions])
        target_q_values = rewards + (1 - dones) * 0.99 * target_critic_values



        with tf.GradientTape() as tape:
            critic_values = self.critic([states, actions])
            critic_loss = tf.reduce_mean(tf.square(target_q_values - critic_values))
        critic_grad = tape.gradient(critic_loss, self.critic.trainable_variables)
        self.critic.optimizer.apply_gradients(zip(critic_grad, self.critic.trainable_variables))




        with tf.GradientTape() as tape:
            actions_pred = self.actor(states)
            critic_values_pred = self.critic([states, actions_pred])

            actor_loss = -tf.math.reduce_mean(critic_values_pred)
        actor_grad = tape.gradient(actor_loss, self.actor.trainable_variables)
        self.actor.optimizer.apply_gradients(zip(actor_grad, self.actor.trainable_variables))



    def act(self, states):

        actions = self.actor(states)

        clipped_actions = tf.clip_by_value(actions, -1, 1)
        return clipped_actions


    def update_target(self):

        actor_weights = self.actor.get_weights()
        target_actor_weights = self.target_actor.get_weights()
        for i in range(len(actor_weights)):
            target_actor_weights[i] = tau * actor_weights[i] + (1 - tau) * target_actor_weights[i]
        self.target_actor.set_weights(target_actor_weights)


        critic_weights = self.critic.get_weights()
        target_critic_weights = self.target_critic.get_weights()
        for i in range(len(critic_weights)):
            target_critic_weights[i] = tau * critic_weights[i] + (1 - tau) * target_critic_weights[i]
        self.target_critic.set_weights(target_critic_weights)




env = CommunicationEnvironment(max_steps=100)
state = env.reset()
state_size = len(state)
action_size = len(env.Phi)

tau = 0.001



agent = DDPGAgent(state_size, action_size, actor_lr=0.001, critic_lr=0.002, buffer_size=100000, batch_size=64, tau=tau)


agent = DDPGAgent(state_size, action_size, actor_lr=0.001, critic_lr=0.002, buffer_size=100000, batch_size=64, tau=tau)

energy_harvested_per_episode = []
transmit_powers = []
phase_shifts = []
uav_positions = []


for episode in range(100):
    state = env.reset()
    total_reward = 0
    done = False

    while not done:
        action = agent.act(np.array([state]))[0]
        next_state, reward, done = env.step(action)
        agent.replay_buffer.add(state, action, reward, next_state, done)
        agent.learn()
        state = next_state
        total_reward += reward

    agent.update_target_networks()
    print(f"Episode: {episode}, Total Reward: {total_reward}")


    energy_harvested_per_episode.append(env.get_energy_harvested())
    uav_positions.append(env.get_uav_position())
    if episode == 99:
      transmit_powers = env.get_transmit_power()
      phase_shifts = env.get_phase_shifts()
# Display the results in tabular format
print("\nEnergy Harvested Per Episode:")
print(tabulate([[i, eh] for i, eh in enumerate(energy_harvested_per_episode)], headers=["Episode", "Energy Harvested"], tablefmt="grid"))

print("\nTransmit Powers for Each User:")
print(tabulate([[i, tp] for i, tp in enumerate(transmit_powers)], headers=["User", "Transmit Power"], tablefmt="grid"))

print("\nPhase Shifts:")
print(tabulate([[i, ps] for i, ps in enumerate(phase_shifts)], headers=["Element", "Phase Shift"], tablefmt="grid"))

print("\nUAV Positions:")
print(tabulate([[i] + list(pos) for i, pos in enumerate(uav_positions)], headers=["Episode", "X-Coordinate", "Y-Coordinate", "Z-Coordinate"], tablefmt="grid"))


max_energy_harvested = max(energy_harvested_per_episode)
energy_harvested_ratio = [eh / max_energy_harvested for eh in energy_harvested_per_episode]




transmit_powers_db = 10 * np.log10(transmit_powers)


print("\nTransmit Powers for Each User in dB (Last Episode):")
print(tabulate([[f"User {i+1}", tp_db] for i, tp_db in enumerate(transmit_powers_db)], headers=["User", "Transmit Power (dB)"], tablefmt="grid"))
```

```python
import numpy as np
import random
import matplotlib.pyplot as plt


N = 800
K = 3
T = 60


base_station_position = np.array([-200, 0, 0])
uav_initial_position = np.array([100, 50, 100])
obstacle_position = np.array([80, 40, 0])


Pk_initial = np.array([0.1, 0.2, 0.3])  # Initial power levels for users
channel_gain_initial = np.array([88.716117, 28.145844, 148.362561])  # Initial channel gains
noise_variance = 10**(-8)

# Genetic algorithm parameters
population_size = 50
num_generations = 100
mutation_rate = 0.1
crossover_rate = 0.7

# Initialize an individual with constraints
def initialize_individual():
    power_allocation = np.random.rand(K) * (Pk_initial * 2)  # Ensure positive power levels
    uav_position = uav_initial_position + np.random.uniform(-10, 10, 3)  # UAV position
    uav_position = np.clip(uav_position, [0, 0, 0], [200, 100, 200])  # Avoid obstacle
    ris_phase_shifts = np.exp(1j * 2 * np.pi * np.random.rand(N))


    ris_grouping = np.random.choice(['info', 'energy'], size=N)
    return {'power_allocation': power_allocation, 'uav_position': uav_position, 'ris_phase_shifts': ris_phase_shifts, 'ris_grouping': ris_grouping}

# Mutation with bounds
def mutate(individual):
    if random.random() < mutation_rate:
        individual['power_allocation'] += np.random.normal(0, 0.05, K)
        individual['power_allocation'] = np.clip(individual['power_allocation'], 0, None)
        individual['uav_position'] += np.random.uniform(-1, 1, 3)
        individual['uav_position'] = np.clip(individual['uav_position'], [0, 0, 0], [200, 100, 200])
        phase_noise = np.exp(1j * 2 * np.pi * np.random.normal(0, 0.05, N))
        individual['ris_phase_shifts'] *= phase_noise
    return individual

# Fitness function with improved channel model and energy harvesting
def fitness_function(individual):
    power_allocation = individual['power_allocation']
    uav_position = individual['uav_position']
    ris_grouping = individual['ris_grouping']

    path_loss = np.linalg.norm(uav_position - base_station_position)
    channel_gain = channel_gain_initial / path_loss**2
    snr = (channel_gain / noise_variance) * power_allocation

    # Calculate energy efficiency and harvested energy
    energy_transmitted = power_allocation * T
    energy_harvested = np.sum(ris_grouping == 'energy') * channel_gain * T
    energy_efficiency = np.sum(snr) / np.sum(energy_transmitted)

    # Penalize if the UAV is too close to the obstacle
    obstacle_penalty = np.exp(-np.linalg.norm(uav_position - obstacle_position))

    fitness = (energy_efficiency + np.sum(energy_harvested)) * obstacle_penalty
    return fitness

# Selection using tournament selection
def select(population, fitness_scores, tournament_size=3):
    selected_individuals = []
    for _ in range(len(population)):
        participants = random.choices(list(zip(population, fitness_scores)), k=tournament_size)
        winner = max(participants, key=lambda x: x[1])
        selected_individuals.append(winner[0])
    return selected_individuals

def calculate_total_energy_harvested(individual):
    ris_grouping = individual['ris_grouping']
    energy_harvested = np.sum(ris_grouping == 'energy') * channel_gain_initial.mean() * T
    return energy_harvested
# Crossover operation
def crossover(parent1, parent2):
    child = {
        'power_allocation': (parent1['power_allocation'] + parent2['power_allocation']) / 2,
                # Crossover operation
        'uav_position': (parent1['uav_position'] + parent2['uav_position']) / 2,
        'ris_phase_shifts': (parent1['ris_phase_shifts'] + parent2['ris_phase_shifts']) / 2,
        'ris_grouping': [np.random.choice([g1, g2]) for g1, g2 in zip(parent1['ris_grouping'], parent2['ris_grouping'])]
    }
    return child



# Running the genetic algorithm
def run_genetic_algorithm():
    population = [initialize_individual() for _ in range(population_size)]
    best_fitness_history = []

    for generation in range(num_generations):

        fitness_scores = [fitness_function(individual) for individual in population]

        best_fitness = max(fitness_scores)
        best_fitness_history.append(best_fitness)


        selected_individuals = select(population, fitness_scores)
        next_generation = []


        while len(next_generation) < population_size:
            parents = random.sample(selected_individuals, 2)
            child = crossover(parents[0], parents[1])
            child = mutate(child)
            next_generation.append(child)


        population = next_generation

    best_individual = max(population, key=fitness_function)


    return best_individual, best_fitness_history


best_solution, fitness_histor = run_genetic_algorithm()

# Plotting the fitness improvement
plt.plot(fitness_history, label='Best Fitness')
plt.xlabel('Generation')
plt.ylabel('Fitness')
plt.title('Best Fitness Over Generations')
plt.legend()
plt.show()





```

```python
#Implement swipt+Phase  Shift+UAV  position using DDPG Algorithm  with NoMA  for  100  episodes.
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Dense, Input, BatchNormalization, Activation
from tensorflow.keras.optimizers import Adam
import random
import collections
from tabulate import tabulate

class CommunicationEnvironment:
    def __init__(self, max_steps):

        self.sinr_history = []
        self.efficiency_history = []
        self.stability_check_window = 10


        self.current_step = 0
        self.max_steps = max_steps
        # Channel gains
        self.H1 = np.random.rand(1, 1)
        self.h2k = np.random.rand(800, 3)


        self.R = np.random.uniform(0, 30)
        self.alpha = np.random.uniform(2, 5)
        self.beta = np.random.uniform(0, 1)


        self.M = 1
        self.N = 800
        self.K = 3
        self.power_for_energy_harvesting = np.zeros(self.K)



        self.Phi = np.exp(1j * np.random.rand(self.N) * 2 * np.pi)
        self.wi = np.random.rand(self.M, self.K)
        self.x = np.random.rand(self.M, 1)
        self.sk = np.random.rand(self.K, 1)
        self.sigma = 10**(-8/10)
        self.gamma_k = np.random.rand(self.K)
        self.B = 1e6
        self.zeta = 1.2
        self.P_BS = 10**(39/10)
        self.P_k = np.random.rand(self.K) * 10**(30/10)
        self.P_UAV = 10**(30/10)
        self.P_UE = 10**(10/10)
        self.P_RIS = self.N * 10**(0.8/10)
        self.P_RISn = 10**(0.8/10)


        self.L_BS = np.array([-200, 0, 0])
        self.Q_j = np.array([100, 50, 100])
        self.user_positions = np.array([[75, 25, 0], [125, 25, 0], [100, 75, 0]])
        self.obstacle_position = np.array([80, 40, 0])




    def reset(self):

        self.H1 = np.random.rand(1, 1)
        self.h2k = np.random.rand(800, 3)


        self.Q_j = np.array([100, 50, 100])




        self.Phi = np.exp(1j * np.random.rand(self.N) * 2 * np.pi)
        self.wi = np.random.rand(self.M, self.K)
        self.x = np.random.rand(self.M, 1)
        self.sk = np.random.rand(self.K, 1)
        self.gamma_k = np.random.rand(self.K)



        state = self._get_state()
        return state

    def _get_state(self):


        state = np.concatenate((self.H1.flatten(), self.h2k.flatten(), self.Q_j, self.user_positions.flatten()))
        return state


    def step(self, action):


        self._apply_action(action)

        previous_energy_harvested = self.get_energy_harvested()
        self._update_environment()


        self.adjust_uav_position_for_optimal_energy_harvesting()
        # Include NOMA power allocation in the step
        self._noma_power_allocation()




        reward = self._calculate_reward()


        done = self._check_done()


        new_state = self._get_state()

        return new_state, reward, done


    def adjust_uav_position_for_optimal_energy_harvesting(self):

        best_position = self.Q_j[:2]
        best_energy = self.get_energy_harvested()
        best_h2k = np.copy(self.h2k)

        for _ in range(10):
            trial_position = self.Q_j[:2] + (np.random.rand(2) - 0.5) * 10
            trial_position = self._apply_position_constraints(trial_position)


            original_position = np.copy(self.Q_j)
            self.Q_j[:2] = trial_position


            self._update_environment()


            trial_energy = self.get_energy_harvested()


            if trial_energy > best_energy:
                best_position = trial_position
                best_energy = trial_energy
                best_h2k = np.copy(self.h2k)


            self.Q_j = original_position


        self.Q_j[:2] = best_position
        self.h2k = best_h2k
        self._update_environment()










    def _apply_action(self, action):
      uav_movement = action[:2]
      phase_shift_changes = action[2:]
      new_position = self.Q_j[:2] + uav_movement
      self.Q_j[:2] = self._apply_position_constraints(new_position)


      if len(phase_shift_changes) < len(self.Phi):
        phase_shift_changes = np.concatenate([phase_shift_changes, np.zeros(len(self.Phi) - len(phase_shift_changes))])
      elif len(phase_shift_changes) > len(self.Phi):
        phase_shift_changes = phase_shift_changes[:len(self.Phi)]

    # Convert phase_shift_changes to a numpy array if it is not already
      phase_shift_changes = np.array(phase_shift_changes)

    # Extract the current phase angles from self.Phi
      current_angles = np.angle(self.Phi)

    # Apply the phase shift changes and wrap them within [0, 2π]
      new_angles = (current_angles + phase_shift_changes) % (2 * np.pi)

    # Preserve magnitudes and update phases of self.Phi
      self.Phi = np.abs(self.Phi) * np.exp(1j * new_angles)









    def _apply_position_constraints(self, position):

        max_x, max_y = 200, 200
        min_x, min_y = -200, -200
        constrained_x = np.clip(position[0], min_x, max_x)
        constrained_y = np.clip(position[1], min_y, max_y)
        return np.array([constrained_x, constrained_y])

    def _update_environment(self):




        distance_bs_ris = np.linalg.norm(self.L_BS - self.Q_j)
        self.H1 = self._calculate_path_loss(distance_bs_ris)


        for k in range(self.K):
            distance_ris_user = np.linalg.norm(self.user_positions[k][:2] - self.Q_j[:2])
            self.h2k[:, k] = self._calculate_path_loss(distance_ris_user) * self._apply_phase_shifts(self.Phi)


        self._update_swipt_signals()










    def _calculate_path_loss(self, distance):
      frequency = 2.4e9
      c = 3e8
      path_loss_0 = 20 * np.log10(4 * np.pi * frequency / c)

      path_loss = 20 * np.log10(distance) + path_loss_0
      path_loss_linear = 10 ** (-path_loss / 10)

      return path_loss_linear



    def _apply_phase_shifts(self, phi):

        return np.exp(1j * phi)

    def _update_swipt_signals(self):

        power_split_ratio = 0.5
        self.power_for_information = self.P_k * power_split_ratio
        self.power_for_energy_harvesting = self.P_k * (1 - power_split_ratio)


    def _calculate_reward(self):

        sinr_values = []
        sorted_indices = np.argsort(self.h2k.mean(axis=0))
        for k in sorted_indices:
            signal_power = np.abs(self.h2k[:, k])**2 * self.P_k[k]
            interference_power = np.sum([np.abs(self.h2k[:, i])**2 * self.P_k[i] for i in sorted_indices if i > k])
            noise_power = self.sigma
            sinr = signal_power / (interference_power + noise_power)
            sinr_values.append(sinr)

        # Calculate energy efficiency
        energy_efficiency = self._calculate_energy_efficiency()

        # Combine the SINR and energy efficiency to form the reward
        reward_signal_quality = np.log(1 + np.mean(sinr_values))
        total_reward = reward_signal_quality + energy_efficiency
        return total_reward

    def _calculate_signal_quality(self):


        sinr_values = []

        for k in range(self.K):


            signal_power = np.abs(self.h2k[:, k])**2


            interference_power = 1


            noise_power = self.sigma


            sinr = signal_power / (interference_power + noise_power)
            sinr_values.append(sinr)


        average_sinr = np.mean(sinr_values)


        signal_quality_reward = np.log(1 + average_sinr)

        self.sinr_history.append(signal_quality_reward)


        return signal_quality_reward


    def _calculate_energy_efficiency(self):

        total_energy_harvested = np.sum(self.power_for_energy_harvesting)


        total_energy_used = np.sum(self.power_for_information)


        if total_energy_used > 0:
            energy_efficiency = total_energy_harvested / total_energy_used
        else:
            energy_efficiency = 0


        energy_efficiency_reward = energy_efficiency

        self.efficiency_history.append(energy_efficiency_reward)

        return energy_efficiency_reward

    def _check_done(self):

        self.current_step += 1
        if self.current_step >= self.max_steps:
            return True
        return False

    def _check_stability_or_convergence(self):

        sinr_threshold = 0.01
        efficiency_threshold = 0.01


        if len(self.sinr_history) >= self.stability_check_window:
            recent_sinr_changes = np.abs(np.diff(self.sinr_history[-self.stability_check_window:]))
            recent_efficiency_changes = np.abs(np.diff(self.efficiency_history[-self.stability_check_window:]))


            if np.all(recent_sinr_changes < sinr_threshold) and np.all(recent_efficiency_changes < efficiency_threshold):
                return True

        return False

    def get_energy_harvested(self):

        return np.sum(self.power_for_energy_harvesting)

    def get_transmit_power(self):

        return self.P_k

    def get_uav_position(self):

        return self.Q_j

    def get_phase_shifts(self):

        return np.angle(self.Phi)

    def _noma_power_allocation(self):

        sorted_indices = np.argsort(self.h2k.mean(axis=0))

        total_power = self.P_BS
        power_allocations = np.zeros(self.K)
        for idx in sorted_indices:

            power_allocations[idx] = total_power / (self.K - idx)
            total_power -= power_allocations[idx]
        self.P_k = power_allocations





class Actor(Model):

    def __init__(self, state_size, action_size, learning_rate=0.001):
        super(Actor, self).__init__()
        self.state_size = state_size
        self.action_size = action_size
        self.learning_rate = learning_rate


        self.dense1 = Dense(400, input_dim=self.state_size, activation='relu')
        self.dense2 = Dense(300, activation='relu')
        self.output_layer = Dense(self.action_size, activation='tanh')  # 'tanh' for bounded actions

        self.compile(optimizer=Adam(learning_rate=self.learning_rate))

    def call(self, state):
        x = self.dense1(state)
        x = self.dense2(x)
        action = self.output_layer(x)
        return action



class Critic(Model):

    def __init__(self, state_size, action_size, learning_rate=0.001):
        super(Critic, self).__init__()
        self.state_size = state_size
        self.action_size = action_size
        self.learning_rate = learning_rate


        self.state_input = Input(shape=(self.state_size,))
        self.state_h1 = Dense(400, activation='relu')(self.state_input)
        self.state_h2 = Dense(300, activation='relu')(self.state_h1)


        self.action_input = Input(shape=(self.action_size,))
        self.action_h1 = Dense(300, activation='relu')(self.action_input)


        merged = tf.keras.layers.Concatenate()([self.state_h2, self.action_h1])


        self.output_layer = Dense(1, activation='linear')(merged)

        self.model = Model(inputs=[self.state_input, self.action_input], outputs=self.output_layer)
        self.optimizer = Adam(learning_rate=self.learning_rate)
        self.model.compile(loss='mse', optimizer=self.optimizer)

    def call(self, inputs):
        return self.model(inputs)

class ReplayBuffer:
    def __init__(self, buffer_size):
        self.buffer_size = buffer_size
        self.buffer = collections.deque(maxlen=buffer_size)

    def add(self, state, action, reward, next_state, done):
        experience = (state, action, reward, next_state, done)
        self.buffer.append(experience)

    def sample(self, batch_size):
        batch = random.sample(self.buffer, min(len(self.buffer), batch_size))
        states, actions, rewards, next_states, dones = map(np.stack, zip(*batch))
        return states, actions, rewards, next_states, dones

class DDPGAgent:
    def __init__(self, state_size, action_size, actor_lr, critic_lr, buffer_size, batch_size, tau=0.001):
        self.state_size = state_size
        self.action_size = action_size
        self.buffer_size = buffer_size
        self.batch_size = batch_size
        self.tau = tau


        self.actor = Actor(state_size, action_size, actor_lr)
        self.critic = Critic(state_size, action_size, critic_lr)


        self.target_actor = Actor(state_size, action_size, actor_lr)
        self.target_critic = Critic(state_size, action_size, critic_lr)


        self.replay_buffer = ReplayBuffer(buffer_size)


        self.update_target_networks()

    def update_target_networks(self):

        self._soft_update(self.actor, self.target_actor)
        self._soft_update(self.critic, self.target_critic)

    def _soft_update(self, model, target_model):
        for target_param, param in zip(target_model.trainable_variables, model.trainable_variables):
            target_param.assign(self.tau * param + (1 - self.tau) * target_param)

    def learn(self):

        states, actions, rewards, next_states, dones = self.replay_buffer.sample(self.batch_size)


        target_actions = self.target_actor(states)
        target_critic_values = self.target_critic([next_states, target_actions])
        target_q_values = rewards + (1 - dones) * 0.99 * target_critic_values



        with tf.GradientTape() as tape:
            critic_values = self.critic([states, actions])
            critic_loss = tf.reduce_mean(tf.square(target_q_values - critic_values))
        critic_grad = tape.gradient(critic_loss, self.critic.trainable_variables)
        self.critic.optimizer.apply_gradients(zip(critic_grad, self.critic.trainable_variables))




        with tf.GradientTape() as tape:
            actions_pred = self.actor(states)
            critic_values_pred = self.critic([states, actions_pred])
            # We want to maximize the critic's predicted value
            actor_loss = -tf.math.reduce_mean(critic_values_pred)
        actor_grad = tape.gradient(actor_loss, self.actor.trainable_variables)
        self.actor.optimizer.apply_gradients(zip(actor_grad, self.actor.trainable_variables))



    def act(self, states):

        actions = self.actor(states)

        clipped_actions = tf.clip_by_value(actions, -1, 1)
        return clipped_actions


    def update_target(self):

        actor_weights = self.actor.get_weights()
        target_actor_weights = self.target_actor.get_weights()
        for i in range(len(actor_weights)):
            target_actor_weights[i] = tau * actor_weights[i] + (1 - tau) * target_actor_weights[i]
        self.target_actor.set_weights(target_actor_weights)


        critic_weights = self.critic.get_weights()
        target_critic_weights = self.target_critic.get_weights()
        for i in range(len(critic_weights)):
            target_critic_weights[i] = tau * critic_weights[i] + (1 - tau) * target_critic_weights[i]
        self.target_critic.set_weights(target_critic_weights)




env = CommunicationEnvironment(max_steps=100)
state = env.reset()
state_size = len(state)
action_size = len(env.Phi)

tau = 0.001
agent = DDPGAgent(state_size, action_size, actor_lr=0.001, critic_lr=0.002, buffer_size=100000, batch_size=64, tau=tau)
agent = DDPGAgent(state_size, action_size, actor_lr=0.001, critic_lr=0.002, buffer_size=100000, batch_size=64, tau=tau)

energy_harvested_per_episode = []
transmit_powers = []
phase_shifts = []
uav_positions = []


for episode in range(100):
    state = env.reset()
    total_reward = 0
    done = False

    while not done:
        action = agent.act(np.array([state]))[0]
        next_state, reward, done = env.step(action)
        agent.replay_buffer.add(state, action, reward, next_state, done)
        agent.learn()
        state = next_state
        total_reward += reward

    agent.update_target_networks()
    print(f"Episode: {episode}, Total Reward: {total_reward}")


    energy_harvested_per_episode.append(env.get_energy_harvested())
    uav_positions.append(env.get_uav_position())
    if episode == 99:
      transmit_powers = env.get_transmit_power()
      phase_shifts = env.get_phase_shifts()

print("\nEnergy Harvested Per Episode:")
print(tabulate([[i, eh] for i, eh in enumerate(energy_harvested_per_episode)], headers=["Episode", "Energy Harvested"], tablefmt="grid"))
print("\nPhase Shifts:")
print(tabulate([[i, ps] for i, ps in enumerate(phase_shifts)], headers=["Element", "Phase Shift"], tablefmt="grid"))
print("\nTransmit Powers for Each User:")
print(tabulate([[f"User {i+1}", tp] for i, tp in enumerate(env.P_k)], headers=["User", "Transmit Power"], tablefmt="grid"))
print("\nUAV Positions:")
print(tabulate([[i] + list(pos) for i, pos in enumerate(uav_positions)], headers=["Episode", "X-Coordinate", "Y-Coordinate", "Z-Coordinate"], tablefmt="grid"))
max_energy_harvested = max(energy_harvested_per_episode)
energy_harvested_ratio = [eh / max_energy_harvested for eh in energy_harvested_per_episode]
transmit_powers_db = 10 * np.log10(transmit_powers)
print("\nTransmit Powers for Each User in dB (Last Episode):")
print(tabulate([[f"User {i+1}", tp_db] for i, tp_db in enumerate(transmit_powers_db)], headers=["User", "Transmit Power (dB)"], tablefmt="grid"))
```
